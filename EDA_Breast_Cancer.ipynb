```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("Breast_Cancer.csv")
```

```python
df.head()
#df.tail()
```

```python
df.columns=df.columns.str.replace(' ','_')
df.rename(columns={'T_Stage_':'T_Stage'},inplace=True)
```

```python
num_cols = df.dtypes[df.dtypes != "object"].index.tolist()
for num_col in num_cols:
  sns.displot(df[num_col], kde=True)
  plt.show()
```

```python
sns.pairplot(df,hue='Status')
```

```python
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation between numerical features')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title('Number of Alive Vs Number of Dead')
ax = sns.countplot(df['Status'], data=df)
total = float(len(df['Status']))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height+5, '{:.1f}%'.format(height/total*100), ha="center")

plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title('Number of Divided by Race')
sns.countplot(data = df, x="Race") 
```

```python
df['differentiate'].value_counts()
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(df['differentiate'], hue=df['Status'])
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.xticks(rotation = 30)
plt.xlabel("Differentiation Status",fontdict={'fontsize': 15})
plt.ylabel("Count",fontdict={'fontsize': 15})
plt.title("Dead/Alive based on Differentiation Status",fontdict={'fontsize': 20},
          fontweight="bold")
```

```python
df['6th_Stage'].value_counts()
```

```python
order_stage = ['IIA', 'IIB', 'IIIA', 'IIIB', 'IIIC']

plt.figure(figsize=(5, 4), dpi=100)
sns.countplot(data=df, x='6th_Stage', hue='Status', order=order_stage)
plt.ylabel('Values')
plt.title('Dead/Alive based on different stage groups');
```

```python
df.Grade.value_counts()
```

```python
count_cols = ['Survival_Months', 'Age', 'Tumor_Size']
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10,8))
fig.suptitle('Relationship of Cancer Grade with Survival Months, Age and Tumor Size');
for i, col in enumerate(count_cols):
    sns.boxplot(data=df, x='Grade', y=col, ax=axes[i])
    

plt.tight_layout()
plt.show()
```

```python
count_cols = ['Race', 'Marital_Status', 'T_Stage', 'N_Stage']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
axes = axes.flatten()

for i, col in enumerate(count_cols):
    sns.countplot(x=col, data=df, hue='Status', ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```
